```python
import requests
import time

res = requests.get("http://api.open-notify.org/iss-now.json")

data = []
data_points = 105
for i in range(data_points):
    print(i)
    res = requests.get("http://api.open-notify.org/iss-now.json")
    json_data = res.json()
    data.append(json_data)
    time.sleep(10)
```

```python
data
```

```python
import matplotlib.pyplot as plt

body = []
for i in data:
    row = []
    row.append(i["timestamp"])
    row.append(i["iss_position"]["latitude"])
    row.append(i["iss_position"]["longitude"])
    body.append(row)
    
latitude = [float(j[1]) for j in body]
longitude = [float(j[2]) for j in body]

plt.figure(figsize = (10,10))
plt.plot(longitude, latitude, '-o', label='ISS Position', color='beige')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('International Space Station Position')
plt.legend()
plt.grid(True)
plt.show()
```
